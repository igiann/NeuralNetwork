$$C=(o-y)^2$$
$$\frac{dC}{dw}=\frac{dC}{do}\frac{do}{dw}=2(o-y)x$$

```python
# Simplest neural network
import numpy as np
import matplotlib.pyplot as plt

x = 1.5
w = 0.8
y = 0.4
r = 0.01

c = np.zeros(100)
for i in range(len(c)):
    o = x*w
    w -= r*2*(o-y)*x
    c[i]= (o-y)**2

print(w)
plt.plot(c)
plt.show()
```

Step one:
$$o_{temp} = x*w$$
Step two:
$$o=\frac{1}{1+e^{-o_{temp}}}$$
Cost function:
$$C=(o-y)^2$$

$$\frac{dC}{dw}=\frac{dC}{do}\frac{do}{do_{temp}}\frac{do_{temp}}{dw}=2(o-y)o(1-o)x$$

```python
# Simplest neural network
import numpy as np
import matplotlib.pyplot as plt

x = 1.5
w = 0.8
y = 0.4
r = 0.1

c = np.zeros(500)
for i in range(len(c)):
    o_temp = x*w
    o = 1/(1+np.exp(-o_temp))
    w -= r*2*(o-y)*o*(1-o)*x
    c[i]= (o-y)**2

plt.plot(c)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x  = np.array([[0.5,0.7]])
wh = np.array([[0.1,0.2],[0.3,0.4]])
wo = np.array([[0.1],[0.2]])
r = 0.1
y = 0.5
c=np.zeros(1000)
for i in range(len(c)):
    h_temp = x@wh
    h = 1/(1+np.exp(-h_temp))
    o_temp = h@wo
    o = 1/(1+np.exp(-o_temp))
    wo -= r*2*(o-y)*o*(1-o)*h.T
    wh -= r*2*(o-y)*o*(1-o)*h*(1-h)*x
    c[i]= (o-y)**2
plt.plot(c)
plt.show()
print(wh)
```
